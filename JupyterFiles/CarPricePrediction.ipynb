```python
x_car = car_col
y_car = car['price']
x=car.drop(['price'],axis=1) 
y = car['price']
```

```python
from sklearn.preprocessing import StandardScaler
sc_car=StandardScaler()
sc_x_car= sc_car.fit_transform(x_car)

sc=StandardScaler()
sc_x=sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train_car,x_test_car,y_train_car,y_test_car=train_test_split(sc_x_car,y_car,test_size=0.20,random_state=401)

x_train,x_test,y_train,y_test=train_test_split(sc_x,y,test_size=0.80,random_state=401)
```

```python
from sklearn.linear_model import LinearRegression
lr_car=LinearRegression()
lr_car.fit(x_train_car,y_train_car)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred_lr_car=lr_car.predict(x_test_car)

y_pred_lr=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test_car,y_pred_lr_car))
print('*'*20)
print(r2_score(y_test,y_pred_lr))
```

```python
from sklearn.metrics import mean_squared_error
mse_lr_car=mean_squared_error(y_test_car,y_pred_lr_car)
mse_lr=mean_squared_error(y_test,y_pred_lr)
print('Mean squared error without feature selection: {}'.format(np.round(mse_lr_car,2)))
print('Mean squared error with feature selection: {}'.format(np.round(mse_lr,2)))
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
x_train,x_test,y_train,y_test=train_test_split(sc_x,y,test_size=0.20,random_state=500)
```

```python
DNN=tf.keras.models.Sequential()
DNN.add(tf.keras.layers.Dense(units=20,activation='relu'))
DNN.add(tf.keras.layers.Dense(units=20,activation='relu'))
DNN.add(tf.keras.layers.Dense(units=1))
DNN.compile(optimizer='adam',loss='mean_squared_error')
DNN.fit(x_train,y_train,batch_size=32,epochs=500)
```

```python
y_pred_DNN=DNN.predict(x_test)
```

```python
print(r2_score(y_test,y_pred_DNN))
```
